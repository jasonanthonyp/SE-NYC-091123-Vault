```python
class Transaction:
    counter, catalog = 0, []

    def __init__(self, customer, coffee, price):
        self.customer = customer
        self.coffee = coffee
        self.price = price

        Transaction.counter += 1
        Transaction.catalog.append(self)

    def __repr__(self):
        return f"{self.customer.name} ordered a {self.coffee.name} for ${self.price}."

    @property
    def price(self):
        self._price = price

    @price.setter
    def price(self, price):
        if type(price) in (int, float) and (1 <= price <= 10):
            self._price = price
        else:
            raise Exception("girl what?")

    @property
    def customer(self):
        self._customer = customer

    @customer.setter
    def customer(self, customer):
        if isinstance(customer, Customer):
            self._customer = customer
        else: 
            raise Exception("hate this `transaction.order` ")

    @property
    def coffee(self):
        self._coffee = coffee

    @coffee.setter
    def coffee(self, coffee):
        if isinstance(coffee, Coffee):
            self._coffee = coffee

        else:
            raise Exception("woah now")
            
```
